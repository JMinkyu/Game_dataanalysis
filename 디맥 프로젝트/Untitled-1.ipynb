```python
import cv2
import os

video_path = "DIE IN 5B SC15.mp4"

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("‚ùå ÏòÅÏÉÅ ÌååÏùºÏùÑ Ïó¥ Ïàò ÏóÜÏäµÎãàÎã§. Í≤ΩÎ°úÎ•º ÌôïÏù∏ÌïòÏÑ∏Ïöî.")
    exit()

output_folder = "frames"
os.makedirs(output_folder, exist_ok=True)

frame_number = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("üìå Î™®Îì† ÌîÑÎ†àÏûÑÏùÑ ÏùΩÏóàÍ±∞ÎÇò, ÌîÑÎ†àÏûÑÏùÑ Î∂àÎü¨Ïò§ÏßÄ Î™ªÌñàÏäµÎãàÎã§.")
        break

    print(f"Frame {frame_number}: {type(frame)}")  # ÌîÑÎ†àÏûÑ Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏

    if frame is None:
        print(f"‚ùå ÌîÑÎ†àÏûÑ {frame_number}Ïù¥ ÎπÑÏñ¥ ÏûàÏäµÎãàÎã§. Í±¥ÎÑàÎúÅÎãàÎã§.")
        continue  # Îπà ÌîÑÎ†àÏûÑÏù¥Î©¥ Ï†ÄÏû•ÌïòÏßÄ ÏïäÍ≥† ÎÑòÏñ¥Í∞ê

    # ÌïÑÏöîÌïòÎ©¥ ÏÉâÏÉÅ Î≥ÄÌôò Ï†ÅÏö©
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    frame_filename = os.path.join(output_folder, f"frame_{frame_number:04d}.jpg")
    cv2.imwrite(frame_filename, frame)
    
    frame_number += 1

cap.release()
print("‚úÖ ÌîÑÎ†àÏûÑ Ï∂îÏ∂ú ÏôÑÎ£å!")
```

```python
import numpy as np

# 1. Î∂ÑÏÑùÌï† Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞ (frames Ìè¥ÎçîÏóêÏÑú ÌïòÎÇò ÏÑ†ÌÉù)
image_path = "frames/frame_0393.jpg"
image = cv2.imread(image_path)

if image is None:
    print("‚ùå Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨Ïò¨ Ïàò ÏóÜÏäµÎãàÎã§. Í≤ΩÎ°úÎ•º ÌôïÏù∏ÌïòÏÑ∏Ïöî.")
    exit()

# 2. ROI (ÎÖ∏Ìä∏ ÏòÅÏó≠) ÏÑ§Ï†ï
height, width = image.shape[:2]
roi = image[int(height * 0.27):int(height * 0.55), int(width * 0.35):int(width * 0.65)]  # Ï§ëÏïô Î∂ÄÎ∂ÑÎßå ÏûêÎ¶Ñ

# 3. ÌäπÏ†ï ÏÉâÏÉÅ Î≤îÏúÑ ÏÑ§Ï†ï (Ïòà: ÌååÎûÄÏÉâ ÎÖ∏Ìä∏ Í≤ÄÏ∂ú)
lower_color = np.array([30, 50, 50])   # HSV ÏµúÏÜå Í∞í
upper_color = np.array([1000, 225, 225])  # HSV ÏµúÎåÄ Í∞í

# 4. ÏÉâÏÉÅ ÌïÑÌÑ∞ÎßÅ Ï†ÅÏö©
mask = cv2.inRange(roi, lower_color, upper_color)

# 5. ÌïÑÌÑ∞ÎßÅ Í≤∞Í≥º Ï∂úÎ†•
cv2.imshow("Original Image", roi)  # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ
cv2.imshow("Filtered Mask", mask)  # ÌïÑÌÑ∞ÎßÅÎêú Ïù¥ÎØ∏ÏßÄ
cv2.waitKey(0)
cv2.destroyAllWindows()
```
