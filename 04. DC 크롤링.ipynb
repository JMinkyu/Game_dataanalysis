```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time 
import datetime as dt
```

```python
# 디시인사이드 크롤링
dc_df = pd.DataFrame(columns=['title_num', 'title_text', 'nickname', 'write_time'])

# USER HEADER
headers = [{'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36'}]

#받고싶은 페이지 입력
page = int(input("던갤에서 받고 싶은 페이지수 입력 : "))

for i in range(1, page + 1) :
    join_df = pd.DataFrame(columns=['title_num', 'title_text', 'nickname', 'write_time'])

    titles = []
    nicknames = []
    times = []
    nums = []

    time.sleep(1)
    # url 호출출
    url = f'https://gall.dcinside.com/board/lists/?id=d_fighter_new2&page={i}' # 던전앤파이터 갤러리, 던파아이피 갤러리
    try :
        response = requests.get(url, headers=headers[0])
        print(f'페이지 {i}번 진행중입니다.')
    except :
        print('error')
    soup = BeautifulSoup(response.content, 'html.parser')

    # 데이터 로드
    article_list = soup.find('tbody').find_all('tr')

    for i in range(0, len(article_list)) : 

        # html 전처리
        article_1 = article_list[i]
        title_num = article_1.find("td", {"class" : "gall_num"}).text # 게시글 번호
        if not title_num.isdigit() :
            continue
        title_text = article_1.find_all("a")[0].text.strip()

        try :
            nickname = article_1.find("span", {'class' : "nickname in"}).text
        except AttributeError:
            sub_nickname = article_1.find("span", {'class' : "nickname"})['title']
            sub_ip = article_1.find("span", {'class' : "ip"}).text
            nickname = f"{sub_nickname}{sub_ip}"
        write_time = article_1.find("td", {'class' : 'gall_date'})['title']
        
        nums.append(title_num)
        titles.append(title_text)
        nicknames.append(nickname)
        times.append(write_time)

    join_df = pd.DataFrame({
        'title_num' : nums,
        'title_text' : titles,
        'nickname' : nicknames,
        'write_time' : times
    })
    dc_df = pd.concat([dc_df, join_df])
dc_df = dc_df.set_index(keys = 'title_num').reset_index()
dc_df
```

```python
dc_df['write_time'] = pd.to_datetime(dc_df['write_time'])
```
